```python
import numpy as np
import pandas as pd
```

Read in data, display it

```python
#Read in data
attendance_21 = pd.read_excel('2021Attendance.xlsx',sheet_name='By School')
attendance_22 = pd.read_excel('2022Attendance.xlsx',sheet_name='By School')
# school_names = pd.read_excel("Another School List.xlsx", sheet_name="Closed Schools")
school_names = pd.read_excel("School Names and ID.xlsx")

#Only keep the "attendance rate" for now. In the future we could also include the 'avg days of attendance' or 'avg days of membership'
attendance_21 = attendance_21[attendance_21['Measure'] == 'Attendance Rate']
attendance_22 = attendance_22[attendance_22['Measure'] == 'Attendance Rate']

#rename columns, drop irrelevant columns
attendance_21 = attendance_21.rename(columns={'School_Name':'School Name','school_id':'School ID'})
attendance_21 = attendance_21.drop(columns=['LEA Type','Measure','School Year','LEA_name','district_id','district_nbr','school_number'])
col_names = attendance_21.columns
col_names = [col_names[0]] + [col_names[-1]] + list(col_names[1:-1])
attendance_21 = attendance_21[col_names]

attendance_22 = attendance_22.rename(columns={'School_Name':'School Name','school_id':'School ID'})
attendance_22 = attendance_22.drop(columns=['LEA Type','Measure','School Year','LEA_name','district_id','district_nbr','school_number'])
col_names = attendance_22.columns
col_names = [col_names[0]] + [col_names[-1]] + list(col_names[1:-1])
attendance_22 = attendance_22[col_names]

#rename Utah Arts Academy OLD to Utah Arts Academy in the 2022 data
attendance_22.loc[attendance_22['School Name']=='Utah Arts Academy OLD','School Name'] = 'Utah Arts Academy'

attendance_22
```

Clean up 2021 Data

```python
#only keep high schools that are in the master list
attendance_21_short = attendance_21.copy()
# attendance_21_short = attendance_21[attendance_21['School Name'].str.lower().isin(school_names['School'].str.lower())]
# attendance_21_short_2 = attendance_21[attendance_21['School ID'].isin(school_names['CompositeSchoolID'])]

#Drop data that isn't grades 9-12
attendance_21_short.drop(attendance_21_short.filter(regex='Grade [1-8]$').columns, axis=1,inplace=True)
attendance_21_short.drop(columns=['Kindergarten'],inplace=True)
attendance_21_short.dropna(subset=['Grade 9','Grade 10','Grade 11','Grade 12'],inplace=True)
#Fill any missing attendance data with the "overall" attendance data
for col in attendance_21_short.columns[7:]:
    attendance_21_short[col] = attendance_21_short[col].fillna(attendance_21_short['K-12 Overall'])

#Rename columns
attendance_21_short.rename(columns={'K-12 Overall':'Overall'},inplace=True)
columns_rename = dict()
for col in attendance_21_short.columns[2:]:
    columns_rename[col] = f'{col} Attendance 2021'
attendance_21_short.rename(columns=columns_rename,inplace=True)

attendance_21_short
```

Clean up 2022 Data

```python
#Drop data that isn't grades 9-12
attendance_22_short = attendance_22.copy()
attendance_22_short.drop(attendance_22_short.filter(regex='Grade [1-8]$').columns, axis=1,inplace=True)
attendance_22_short.drop(columns=['Kindergarten'],inplace=True)
attendance_22_short.dropna(subset=['Grade 9','Grade 10','Grade 11','Grade 12'],inplace=True)
#Fill any missing attendance data with the "overall" attendance data
for col in attendance_22_short.columns[7:]:
    attendance_22_short[col] = attendance_22_short[col].fillna(attendance_22_short['K-12 Overall'])

#Rename columns
attendance_22_short.rename(columns={'K-12 Overall':'Overall'},inplace=True)
columns_rename = dict()
for col in attendance_22_short.columns[2:]:
    columns_rename[col] = f'{col} Attendance 2022'
attendance_22_short.rename(columns=columns_rename,inplace=True)

attendance_22_short
```

Combine the 2 into one big dataset

```python
merged_attendance = pd.merge(attendance_21_short,attendance_22_short,'outer',on='School Name')
merged_attendance = merged_attendance.drop(columns=['School ID_x'])

# #inspect the 'extra' schools
# extra_21 = attendance_21_short[~attendance_21_short['School Name'].isin(attendance_22_short['School Name'])]
# extra_22 = attendance_22_short[~attendance_22_short['School Name'].isin(attendance_21_short['School Name'])]


merged_attendance.to_csv('Attendance_Cleaned.csv')
merged_attendance
```
